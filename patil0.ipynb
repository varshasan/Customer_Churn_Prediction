```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
from sklearn.preprocessing import LabelEncoder
from sklearn.cluster import KMeans
```

```python
# data Collection and Analysis 
df=pd.read_csv(r"C:\Users\feedb\Downloads\Mall_Customers.csv")
df.head()
```

```python
#finding theno . of rows and columns 
df.shape 
```

```python
df.info()
```

```python
#checking missing value
df.isnull().sum()
```

```python
label_encoder=LabelEncoder()
df['Gender']=label_encoder.fit_transform(df['Gender'])
```

```python
#chosing the required column
x=df[['Annual Income (k$)', 'Spending Score (1-100)']].values
```

```python
x
```

```python
#choosing number of clusters by using WCSS Formula
wcss=[]
for i in range (1,11):
    kmeans=KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1, len(wcss) + 1), wcss)
plt.title('Elbow point Graph')
plt.xlabel('no.of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
#optimum number of clusters are 4
```

```python
#training of K-means Model
kmeans=KMeans(n_clusters=5, init='k-means++',random_state=42)
# retrn a label for each data point based on there cluster 
y=kmeans.fit_predict(x)
print(y)
```

```python
#visual ization 
plt.figure(figsize=(8,8))

plt.scatter(x[y==0, 0], x[y==0, 1], s=50, label='cluster 1')
plt.scatter(x[y==1, 0], x[y==1, 1], s=50, label='cluster 2')
plt.scatter(x[y==2, 0], x[y==2, 1], s=50, label='cluster 3')
plt.scatter(x[y==3, 0], x[y==3, 1], s=50, label='cluster 4')
plt.scatter(x[y==4, 0], x[y==4, 1], s=50, label='cluster 5')

# centroids
plt.scatter(
    kmeans.cluster_centers_[:, 0],
    kmeans.cluster_centers_[:, 1],
    s=200, marker='x',label='Centroids'
)
plt.title('Customer Groups')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.legend()
plt.show()
    
```
